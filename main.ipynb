## 2. Now write a function that returns the largest element on a list

```python
#your code here
def greatest(input_list):
    '''
    Function accepts a list and returns the largest element
    '''
    max_item = input_list[0]
    for item in input_list:
        if item > max_item:
            max_item = item
    return max_item
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
#your code here
    '''
    Function accepts a number and returns its factorial
    '''
    product = 1
    for i in range(1, n + 1):
        product *= i
    return product
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

```python
# This will test your function 
test_factorial(factorial)
```

```python
def unique(lst_un):
#your code here
    '''Function accepts a list and returns a list of the unique values in that list
    '''
    unique_elements = []
    elements_seen = []
    for i in lst_un:
        if i not in elements_seen:
            unique_elements.append(i)
            elements_seen.append(i) 
    return unique_elements
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. ðŸ˜‰`

```python
def st_dev(list_sd):
#your code here
    '''
    Function accepts a list and returns its standard deviation
    '''
    sum_list = 0
    n = 0
    for i in list_sd:
        sum_list += i
        n += 1
    mean = sum_list / n
    sum_terms = 0
    for i in list_sd:
        sum_terms += (i - mean) ** 2
    print(sum_terms)
    sd = (sum_terms / (n - 1)) ** 0.5
    return sd
        
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def pangram(string):
#your code here
    '''
    Function accepts a string and returns True if it's a pandram, and False if it's not
    '''
    alphabet = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j", "k", "l", "m", "o", "p", "q", "r", "s", "t", "u", "v", "w", "x", "y", "z"]
    str_lower = string.lower()
    for letter in alphabet:
        if letter not in str_lower:
            return False
    return True
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! ðŸ¤ª`

```python
def sort_alpha(string):
#your code here
    '''
    Function accepts a string of comma-separated words and returns a string of comma-separated words sorted alphabetically
    '''
    word_list = []
    current_word = ""
    for char in string:
        if char == ",":
            word_list.append(current_word)
            current_word = ""
        else:
            current_word += char
    word_list.append(current_word)
    sorted_list = sorted(word_list)
    sorted_string = ""
    for i in sorted_list[:-1]:
        sorted_string += i
        sorted_string += ","
    sorted_string += sorted_list[-1]
    return sorted_string
```

```python
# This will test your function 
test_alpha(sort_alpha)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):
#your code here
    '''
    Function accepts a password string, returns True if password contains at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character. It returns False otherwise
    '''
    special_chars = ["!", "@", ",", "#", "$", ".", "[", "]", "(", ")", "{", "}", "*", "%"]
    if len(password) < 8:
        return False
    
    has_upper = False
    has_lower = False
    has_special = False
    has_number = False

    for char in password:
        if char.isupper():
            has_upper = True
        if char.islower():
            has_lower = True
        if char.isnumeric():
            has_number = True
        if char in special_chars:
            has_special = True
    return has_upper and has_lower and has_number and has_special

       
    
```

```python
# This will test your function 
test_pass(check_pass)
```
